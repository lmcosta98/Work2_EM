```python
import pandas as pd
```

```python
data = pd.read_csv('datasets/gdp_vs_happiness.csv')
data.rename(columns={'Life satisfaction in Cantril Ladder (World Happiness Report 2022)': 'Happiness',
             'GDP per capita, PPP (constant 2017 international $)': 'GDP_Per_Capita'}, inplace=True)
data
```

```python
data_2 = pd.read_csv('datasets/Time_Use.csv', sep=';')
data_2
```

```python
countries = ["Australia", "Austria", "Belgium", "Canada", "Denmark", "Estonia", "Finland", "France",
             "Germany", "Greece", "Hungary", "Ireland", "Italy", "Japan", "Korea", "Latvia", "Lithuania",
             "Luxembourg", "Mexico", "Netherlands", "New Zealand", "Norway", "Poland", "Portugal", "Slovenia",
             "Spain", "Sweden", "Turkey", "UK", "USA", "China", "India", "South Africa"]
```

```python
data_hap = data[data.Entity.isin(countries)]
data_hap = data_hap[~data_hap.Happiness.isna()].drop(
    columns=['Code', 'Population (historical estimates)', 'Continent', 'GDP_Per_Capita']).rename(columns={'Entity':'Countries'}).reset_index(drop=True)
data_hap.to_csv('datasets/Happiness.csv', index=False)
```

```python
# Taking only year 2020
data = data[data.Entity.isin(countries)]
data = data[data.Year == 2020].reset_index(drop=True).drop(
    columns=['Population (historical estimates)', 'Continent', 'Code', 'Year'])
data
```

```python
data = data.join(data_2).drop(columns='Country').rename(columns={'Entity': 'Countries'})
data
```

```python
data.to_csv('Time_Happiness_GDP.csv', index=False)
```
